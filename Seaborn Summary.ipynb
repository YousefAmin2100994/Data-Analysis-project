# *Visualization Using Seaborn*

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

## *Categorical Data*

```python
# reading the dataset
data_frame = sns.load_dataset('tips')
 
# first five entries if the dataset
data_frame.head()
```

```python
# put the style first
plt.style.use('ggplot')
```

### *1. Bar Plot*

```python
#A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean.
#It can also be understood as a visualization of the group by action. 
#categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column
sns.barplot(x='sex',y='total_bill',data=data_frame)
plt.xlabel('Sex')
plt.ylabel('Total_Bill')
plt.title('Bar Plot for total_bill vs Sex')
```

### *2.Count Plot*

```python
#A countplot basically counts the categories and returns a count of their occurrences. 
sns.countplot(y='sex',data=data_frame)
```

```python
sns.countplot(x='sex',data=data_frame)
```

### *3.Box Plot*

```python
#A boxplot is sometimes known as the box and whisker plot.It shows the distribution of the quantitative data that represents the comparisons between variables. 
#boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution i.e. 
#the dots indicating the presence of outliers
sns.boxplot(y='total_bill',data=data_frame)
```

```python
sns.boxplot(y='total_bill',x='sex',data=data_frame)
```

```python
sns.boxplot(x='sex',y='total_bill',hue='smoker',data=data_frame)
```

### *4.Violinplot*

```python
#A violin plot depicts distributions of numeric data for one or more groups using density curves. 
#The width of each curve corresponds with the approximate frequency of data points in each region. 
#Densities are frequently accompanied by an overlaid chart type, such as box plot, to provide additional information.
sns.violinplot(y='total_bill',data=data_frame)
```

```python
sns.violinplot(y='total_bill',x='sex',data=data_frame)
```

```python
sns.violinplot(y='total_bill',x='sex',data=data_frame,hue='smoker')
```

### *5.Boxen Plot*

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxenplot(x='total_bill',data=data_frame)
plt.subplot(1,2,2)
sns.boxplot(x='total_bill',data=data_frame)
#use it if there are alot of outliers in the data and to show the data as categories
```

```python
sns.boxenplot(y='total_bill',data=data_frame,x='smoker')
```

## *Numerical Data*

### *1.Dist Plot*

```python
#Dist plot gives us the histogram of the selected continuous variable.
#It is an example of a univariate analysis.
#We can change the number of bins i.e. number of vertical bars in a histogram
sns.distplot(x=data_frame['total_bill'])
```

### *2.Pair Plot*

```python
#It takes all the numerical attributes of the data and plot pairwise scatter plot for two different variables and histograms from the same variables.
sns.pairplot(data_frame)
```
